### Vasicek Model
---

$$ dr(t) = K(\theta - r(t))dt + \sigma d W(t) $$

In the Vasicek model, interest rates can be come negative.

```python
# Vasicek Model

"""
Simulate the interest path taken by a Vasicek model

theta   = mean
k       = rate of reversion to the mean
r0      = initial rate of interest at t=0
t       = period in years
N        = number of intervals for modeling process
seed     = initialisation for Python's random number generator   

"""

import numpy as np

from matplotlib import pyplot as plt

def vasicek(r0, k, theta, sigma, T=1.,N =10, seed = 777):
    np.random.seed = seed
    dt = T/float(N)
    rates =[r0]
    for i in range(N):
        dr = k*(theta-rates[-1])*dt + sigma*np.random.normal()
        rates.append(rates[-1] + dr)
    return range(N+1), rates

# Vasicek returns a list of time periods and interest rates from the Vasicek model

x, y = vasicek(0.01875, 0.20, 0.01, 0.012, 10., 200)

plt.plot(x,y)
plt.show()
```
